```python
# Importing necessary libraries
import cv2
import numpy as np
import matplotlib.pyplot as plt

# Function to display an image
def display_image(title, img):
    plt.figure(figsize=(6,6))
    plt.imshow(cv2.cvtColor(img, cv2.COLOR_BGR2RGB))
    plt.title(title)
    plt.axis('off')
    plt.show()

# 1. Read an image
img = cv2.imread('/content/drive/MyDrive/dorae.png')  # Replace with the path to your image
display_image('Original Image', img)

# 2. Increase the brightness of the image
def increase_brightness(img, value=30):
    # Limit brightness to prevent overflow
    hsv = cv2.cvtColor(img, cv2.COLOR_BGR2HSV)
    h, s, v = cv2.split(hsv)
    v = cv2.add(v, value)
    v = np.clip(v, 0, 255)
    hsv_bright = cv2.merge((h, s, v))
    bright_img = cv2.cvtColor(hsv_bright, cv2.COLOR_HSV2BGR)
    return bright_img

bright_img = increase_brightness(img, value=50)
display_image('Increased Brightness', bright_img)

# 3. Decrease the brightness of the image
def decrease_brightness(img, value=30):
    hsv = cv2.cvtColor(img, cv2.COLOR_BGR2HSV)
    h, s, v = cv2.split(hsv)
    v = cv2.subtract(v, value)
    v = np.clip(v, 0, 255)
    hsv_dark = cv2.merge((h, s, v))
    dark_img = cv2.cvtColor(hsv_dark, cv2.COLOR_HSV2BGR)
    return dark_img

dark_img = decrease_brightness(img, value=50)
display_image('Decreased Brightness', dark_img)
```
