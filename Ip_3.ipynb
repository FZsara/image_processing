```python
import numpy as np
from PIL import Image
import matplotlib.pyplot as plt

# Load image and convert to RGB numpy array
def load_image(image_path):
    image = Image.open(image_path).convert('RGB')
    return np.array(image)

# Monadic operations on each channel
def apply_to_rgb(image_array, operation, *args):
    r, g, b = image_array[:,:,0], image_array[:,:,1], image_array[:,:,2]
    r = operation(r, *args)
    g = operation(g, *args)
    b = operation(b, *args)
    return np.stack([r, g, b], axis=2)

# Monadic operations
def add_constant(image_array, constant):
    return image_array + constant

def subtract_constant(image_array, constant):
    return image_array - constant

def negative(image_array):
    return 255 - image_array

def multiply_constant(image_array, constant):
    return image_array * constant

def divide_into_constant(image_array, constant):
    return constant / image_array

def divide_by_constant(image_array, constant):
    return image_array / constant

def and_operation(image_array, constant):
    return np.bitwise_and(image_array, constant)

def or_operation(image_array, constant):
    return np.bitwise_or(image_array, constant)

def xor_operation(image_array, constant):
    return np.bitwise_xor(image_array, constant)

# Display image using matplotlib
def display_image(image_array, title):
    plt.imshow(np.clip(image_array, 0, 255).astype(np.uint8))
    plt.title(title)
    plt.axis('off')
    plt.show()

# Example usage
if __name__ == "__main__":
    # Load image
    image_path = '/content/drive/MyDrive/dorae.png'
    image_array = load_image(image_path)

    # Display original image
    display_image(image_array, "Original Image")

    # Perform operations (example with constant 50)
    added_image = apply_to_rgb(image_array, add_constant, 50)
    subtracted_image = apply_to_rgb(image_array, subtract_constant, 50)
    negative_image = apply_to_rgb(image_array, negative)
    multiplied_image = apply_to_rgb(image_array, multiply_constant, 1.5)
    divided_into_image = apply_to_rgb(image_array, divide_into_constant, 2000)
    divided_by_image = apply_to_rgb(image_array, divide_by_constant, 2)
    and_image = apply_to_rgb(image_array, and_operation, 50)
    or_image = apply_to_rgb(image_array, or_operation, 50)
    xor_image = apply_to_rgb(image_array, xor_operation, 50)

    # Print and display resulting arrays and images
    operations = {
        "Added 50": added_image,
        "Subtracted 50": subtracted_image,
        "Negative": negative_image,
        "Multiplied by 1.5": multiplied_image,
        "Divided into 2000": divided_into_image,
        "Divided by 2": divided_by_image,
        "AND with 50": and_image,
        "OR with 50": or_image,
        "XOR with 50": xor_image
    }

    for title, image in operations.items():
        print(f"{title}:\n{image}\n")
        display_image(image, title)
```
